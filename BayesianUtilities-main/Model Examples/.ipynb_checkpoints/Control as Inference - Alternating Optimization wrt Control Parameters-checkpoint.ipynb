# Velocity and Acceleration

```julia
A = [1 0.1; 0 1]
B = [0,1]
C = [1 0]

T = 50

# Noisy actuator and sensory data generator
f(x::Vector,u::Real) = MvNormal(A*x+B*u,[0.001/3 0.01/2; 0.01/2 0.1]*0.01)
g(x::Vector) = Normal((C*x)[1],0.00001)

# Initial and terminal points
x_init = 0
x_terminal = 15
y, x_real = zeros(T), zeros(2,T)

# priors
pxs = MvNormal(zeros(2),[0.0001 0;0 0.0001])

# process and emission noise
noise_q = MvNormal(zeros(2),[0.001/3 0.01/2; 0.01/2 0.1])
noise_r = Normal(0,0.1)
;
```

```julia
# Action calculated in an alternating manner
function action(T,state,goal,convergence_mse)
    U = randn(T)
    U_prev = 1*U # Multiplication by 1 prevents pointer behaviour
    mse = 100
    while mse > convergence_mse  
        for t=1:T
            s = goal - A^T*mean(state)
            for i=1:T
                if t!= i
                    s -= A^(T-i)*B*U[i]
                end
            end
            U[t] = (A^(T-t)*B)\(0.5*s)
        end
        mse = sum((U .- U_prev).^2)/length(U)
        U_prev = 1*U
    end
    return U[1]
end
```

```julia
state_estimates = Array{MvNormal}(undef, T)
action_estimates = Array{Real}(undef, T)
;
```

```julia
# Estimate action
u = action(T, pxs, [x_terminal, 0], 1e-10)
action_estimates[1] = u
# Take action in the environment
x_real[:,1] = rand(f(zeros(2),u))
y[1] = rand(g(x_real[:,1]))
# Estimate the state with Kalman filter
predict = A*pxs + B*u + noise_q
state_estimates[1] = collide(predict,C\(y[1]+noise_r))

for t=2:T
    # Estimate action
    u = action(T-t+1, state_estimates[t-1], [x_terminal, 0], 1e-10)
    action_estimates[t] = u
    # Take action in the environment
    x_real[:,t] = rand(f(x_real[:,t-1],u))
    y[t] = rand(g(x_real[:,t]))
    # Estimate the state with Kalman filter
    predict = A*state_estimates[t-1] + B*u + noise_q
    state_estimates[t] = collide(predict,C\(y[t]+noise_r))
end
```
