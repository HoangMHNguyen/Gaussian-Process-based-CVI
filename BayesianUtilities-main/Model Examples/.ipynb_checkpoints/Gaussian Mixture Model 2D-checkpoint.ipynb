```julia
include("../BayesianUtilities.jl")
using Distributions, LinearAlgebra, Plots, Random, .BayesianUtilities
```

```julia
Random.seed!(0);
```

```julia
# Synthetic Data
N = 50 # number of data points

p1 = MvNormal([-10.0, 2.5], [1 0.5;0.5 0.4])
p2 = MvNormal([-3.0, -4.5], [0.25 0; 0 0.5])
p3 = MvNormal([7.5, 1.5], [1 0.2; 0.2 1.25])

real_mix = MixtureModel(MvNormal[p1,p2,p3], [0.5, 0.2, 0.3])

data = rand(real_mix, N);
```

```julia
X = range(-13, 10, length=200)
Y = range(-13, 10, length=200)
Z1 = [pdf(p1, [x,y]) for y in Y, x in X]
Z2 = [pdf(p2, [x,y]) for y in Y, x in X]
Z3 = [pdf(p3, [x,y]) for y in Y, x in X]
contour(X, Y, Z1)
contour!(X, Y, Z2)
contour!(X, Y, Z3)


scatter!(data[1,:], data[2,:], legend=false, color=:black)
```

```julia
# Priors
pm1, pm2, pm3 = MvNormal([-1,1],diagm(0=>10*ones(2))), MvNormal(3*ones(2),diagm(0=>10*ones(2))), MvNormal(-3*ones(2),diagm(0=>10*ones(2)))
pw1, pw2, pw3 = Wishart(2,diagm(0=>ones(2))), Wishart(2,diagm(0=>ones(2))), Wishart(2,diagm(0=>ones(2)))
pϕ = Dirichlet([0.1,0.1,0.1])
;
```

```julia
# Initial recognition factors
qm = [pm1, pm2, pm3]
qw = [pw1, pw2, pw3]
qϕ = pϕ
qz = Array{Categorical}(undef,N)
for n=1:N
    qz[n] = Categorical(3)
end
;
```

```julia
for i=1:20
    # update qz
    for n=1:N
        vmp = normalmix(data[:,n],nothing,qm,qw)
        qz[n] = collide(categorical(nothing,qϕ),vmp)
    end

    # update qϕ
    vmp = categorical.(qz,nothing)
    qϕ = pϕ
    for n=1:N
        qϕ = collide(qϕ,vmp[n])
    end

    # update qm
    qm = [pm1, pm2, pm3]
    for n=1:N
        vmp = normalmix(data[:,n],qz[n],nothing,qw)
        qm = collide.(qm, vmp, canonical=true)
    end
    qm = exp_family.(qm)

    # update qw
    qw = [pw1, pw2, pw3]
    for n=1:N
        vmp = normalmix(data[:,n],qz[n],qm,nothing)
        qw = collide.(qw, vmp, canonical=true)
    end
    qw = exp_family.(qw)
    
    # calculate Free Energy
    F = -entropy(qϕ) - sum(entropy.(qm)) - sum(entropy.(qw))
    F = F + cross_entropy(qϕ, pϕ) + sum(cross_entropy.(qm, [pm1, pm2, pm3])) + sum(cross_entropy.(qw, [pw1, pw2, pw3]))
    for n=1:N
        F -= entropy(qz[n])
        F += cross_entropy(qϕ, qz[n])
        F += normalmix(data[:,n],qz[n],qm,qw)
    end
    @show F
end
```
